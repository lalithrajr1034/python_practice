<h2>Sorting</h2>

```python
import pandas as pd
```

```python
data = pd.read_csv("matches.csv")
```

```python
data.head()
```

```python
(data['team1'].value_counts()+data['team2'].value_counts()).sort_values(ascending=False)
```

```python
data.sort_values('win_by_wickets',ascending=False,inplace=False) 
#inplace = True :: data will change permenantally
```

```python
data.sort_values(['city','date'],ascending=[True, False])# city increase and date decrease
```
