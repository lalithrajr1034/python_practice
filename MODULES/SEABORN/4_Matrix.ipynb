<h1>Matrix plot</h1>

```python
#HEATMAP
#CLUSTERMAP
import seaborn as sns
import plotly as px
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
gap = px.data.gapminder()
gap.head()
```

<h3>Heat map</h3>

```python
#To draw a heatmap we need to create a pivot table 
heat_df = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
sns.heatmap(heat_df)
```

```python
#hence this heatmap is a axis lavel image so we can change a size of this by using matplotlib
plt.figure(figsize = (15,15))
sns.heatmap(heat_df)
```

```python
df = gap[gap['continent'] == 'Europe']
temp_df = df.pivot(index = 'country', columns = 'year', values = 'lifeExp')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot = True, cmap = 'summer', linewidths=1) #summer , rainy
```
