```python
import seaborn as sns
import matplotlib.pyplot as plt
```

figure level ->displot<br>
axis level   ->histplot, kdeplot, rugplot

```python
#Ploting univariant histogram
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.histplot(data = df, x = 'total_bill', bins=20)
```

```python
sns.displot(data = df,x = 'total_bill',kind = 'hist')
```

```python
# we can plot a hisogram in a catogatical column 
sns.displot(data = df, x = 'day')
```

```python
sns.displot(data = df, x = 'tip',hue = 'day')
```

```python
sns.histplot(data = df, x = 'tip', hue = "sex", element = 'step') #element = step #"helps to remove a border of a histplot"
```

```python
titanic_df = sns.load_dataset('titanic')
```

```python
titanic_df.head()
```

```python
sns.histplot(data = titanic_df, x = 'age', hue = "sex", element = 'step')
```

```python
#faciting using col and row -> Only work on a figure lavel function
sns.displot(data = titanic_df, x = 'age', element = 'step', col = 'sex')
```

KDE plot

```python
sns.kdeplot(data = df, x = 'total_bill')
```

```python
sns.displot(data = df, x = 'total_bill', kind = 'kde', hue = 'sex', fill = True)
```

```python
# THIS IS USED WITH OTHER GRAPH AND IT IS USED TO DETECT A DENSITY BY USING A TICKS
sns.kdeplot(data = df, x = 'total_bill', hue = 'sex', fill = True)
sns.rugplot(data = df, x = 'total_bill', hue = 'sex')  
```
