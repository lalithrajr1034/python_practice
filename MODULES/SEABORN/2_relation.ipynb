```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
"""
outer -> Figure
graph -> Axis
"""
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#scatter plot -> axis lavel function
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'smoker', style = "time", size = 'size')
```

```python
#relplot -> figure level -> square shape
sns.relplot(data = tips, x = "total_bill", y = 'tip', kind = 'scatter', hue = 'smoker', style = 'time', size = 'size')
```

```python
#LINE PLOT
```

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

```python
#figure lavel function
sns.relplot(data = temp_df,x ='year', y = 'lifeExp', kind = 'line')
```

```python
temp_df = gap[gap['country'].isin(['India', 'Zimbabwe','China'])]
temp_df
```

```python
#figure
sns.relplot(kind = 'line',data = temp_df, x ='year', y = 'lifeExp', hue = 'country', size = 'continent')
```

```python
#axis lavel plot
sns.lineplot(data = temp_df, x ='year', y = 'lifeExp', hue = 'country', size = 'continent')
```

```python
# Facet plot will work with figure lavel functino
sns.relplot(data = tips, y = 'tip', x = 'total_bill', kind = 'scatter', col = 'sex')
```

```python
sns.relplot(data = tips, y = 'tip', x = 'total_bill', kind = 'scatter', row = 'sex')
```

```python
sns.relplot(data = tips, y = 'tip', x = 'total_bill', kind = 'scatter',row = 'smoker', col = 'sex')
```

```python
sns.relplot(data = tips, y = 'tip', x = 'total_bill', kind = 'scatter',row = 'smoker', col = 'day')
```
